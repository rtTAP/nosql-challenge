### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

degree_search = Decimal('0.01')

# Find Penang Flavours longitude and latitude
penang_flavours = {'BusinessName': 'Penang Flavours'}
penang_flavours_data = establishments.find_one(penang_flavours)
penang_flavours_longitude = penang_flavours_data['geocode']['longitude'].to_decimal()
penang_flavours_latitude = penang_flavours_data['geocode']['latitude'].to_decimal()


# Perform arithmetic operations using Decimal
min_latitude = Decimal128(penang_flavours_latitude - degree_search)
max_latitude = Decimal128(penang_flavours_latitude + degree_search)
min_longitude = Decimal128(penang_flavours_longitude - degree_search)
max_longitude = Decimal128(penang_flavours_longitude + degree_search)

# Construct the query to find top 5 establishments
query = {
    'RatingValue': '5',
    'geocode.latitude': {'$gte': min_latitude, '$lte': max_latitude},
    'geocode.longitude': {'$gte': min_longitude, '$lte': max_longitude}
}

sort = [('scores.Hygiene', 1)]
limit = 5

# Print the results
top_establishments = establishments.find(query).sort(sort).limit(limit)

for result in top_establishments:
    pprint(result)
```

```python
# Convert result to Pandas DataFrame
top_establishments_list = list(establishments.find(query).sort(sort).limit(limit))
top_establishments_df = pd.DataFrame(top_establishments_list)
top_establishments_df
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that:
# 1. Matches establishments with a hygiene score of 0
# 2. Groups the matches by Local Authority
# 3. Sorts the matches from highest to lowest

pipeline = [
    {
        '$match': {
            'scores.Hygiene': 0
        }
    },
    {
        '$group': {
            '_id': '$LocalAuthorityName',
            'count': {'$sum': 1}
        }
    },
    {
        '$sort': {'count': -1}
    }
]

# Print the number of documents in the result

agg_results = list(establishments.aggregate(pipeline))


# Print the first 10 results
pprint(agg_results[:10])
```

```python
# Convert the result to a Pandas DataFrame
agg_df = pd.DataFrame(agg_results)

# Display the number of rows in the DataFrame
print('The number of rows in the dataframe is', len(agg_df))

# Display the first 10 rows of the DataFrame
agg_df.head(10)
```
